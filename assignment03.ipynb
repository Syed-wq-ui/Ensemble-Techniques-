```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import fetch_california_housing # Using California Housing as Boston is deprecated
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import numpy as np

# Load the California Housing dataset
data = fetch_california_housing()
X, y = data.data, data.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# --- 1. Train a single Decision Tree Regressor ---
dt_regressor = DecisionTreeRegressor(random_state=42)
dt_regressor.fit(X_train, y_train)
y_pred_dt = dt_regressor.predict(X_test)
mse_dt = mean_squared_error(y_test, y_pred_dt)
print(f"Decision Tree Regressor MSE: {mse_dt:.4f}")

# --- 2. Train a Random Forest Regressor ---
# n_estimators: The number of trees in the forest. A higher number generally leads to better performance but also increases computation time.
# random_state: Controls the randomness for reproducibility.
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42) #
rf_regressor.fit(X_train, y_train)
y_pred_rf = rf_regressor.predict(X_test)
mse_rf = mean_squared_error(y_test, y_pred_rf)
print(f"Random Forest Regressor MSE: {mse_rf:.4f}")

# --- 3. Compare the performance ---
print("\n--- Performance Comparison ---")
if mse_rf < mse_dt:
    print(f"The Random Forest Regressor performed better with a lower MSE ({mse_rf:.4f}) compared to the Decision Tree Regressor ({mse_dt:.4f}).")
else:
    print(f"The Decision Tree Regressor performed better with a lower MSE ({mse_dt:.4f}) compared to the Random Forest Regressor ({mse_rf:.4f}).")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import numpy as np

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define a list of different numbers of trees to test
n_estimators_list = [10, 50, 100, 200, 500]

# Dictionary to store accuracy scores for each n_estimators value
accuracy_scores = {}

print("Comparing Random Forest Classifier accuracy with different numbers of trees:\n")

for n_estimators in n_estimators_list:
    # Create a Random Forest Classifier with the current number of trees
    rf_classifier = RandomForestClassifier(n_estimators=n_estimators, random_state=42)

    # Train the Random Forest Classifier
    rf_classifier.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = rf_classifier.predict(X_test)

    # Calculate and store the accuracy score
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores[n_estimators] = accuracy

    print(f"Random Forest with n_estimators={n_estimators}: Accuracy = {accuracy:.4f}")

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(list(accuracy_scores.keys()), list(accuracy_scores.values()), marker='o')
plt.title('Random Forest Accuracy vs. Number of Trees')
plt.xlabel('Number of Trees (n_estimators)')
plt.ylabel('Accuracy')
plt.grid(True)
plt.xticks(n_estimators_list)
plt.show()

```

```python
import numpy as np
from sklearn.ensemble import StackingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# --- 1. Train Individual Base Models and get their accuracies ---

# Decision Tree Classifier
dt_classifier = DecisionTreeClassifier(random_state=42)
dt_classifier.fit(X_train, y_train)
y_pred_dt = dt_classifier.predict(X_test)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
print(f"Decision Tree Classifier Accuracy: {accuracy_dt:.4f}")

# Support Vector Machine (SVM) Classifier
# It's good practice to scale data for SVMs
svm_classifier = make_pipeline(StandardScaler(), SVC(random_state=42, probability=True)) # probability=True needed if final_estimator uses predict_proba
svm_classifier.fit(X_train, y_train)
y_pred_svm = svm_classifier.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print(f"SVM Classifier Accuracy: {accuracy_svm:.4f}")

# Logistic Regression Classifier
# It's also good practice to scale data for Logistic Regression
lr_classifier = make_pipeline(StandardScaler(), LogisticRegression(random_state=42, solver='liblinear'))
lr_classifier.fit(X_train, y_train)
y_pred_lr = lr_classifier.predict(X_test)
accuracy_lr = accuracy_score(y_test, y_pred_lr)
print(f"Logistic Regression Classifier Accuracy: {accuracy_lr:.4f}")

# --- 2. Train the Stacking Classifier ---

# Define the base estimators for the Stacking Classifier
# Each estimator is a tuple of (name, estimator instance)
estimators = [
    ('dt', dt_classifier),
    ('svm', svm_classifier),
    ('lr', lr_classifier)
]

# Define the meta-classifier (final_estimator)
# The default final_estimator is LogisticRegression
# Here, we explicitly define it for clarity, and can choose other models as well
final_estimator = LogisticRegression(random_state=42, solver='liblinear')

# Create the Stacking Classifier
# cv=5 ensures cross-validation is used to generate meta-features, preventing data leakage
stacking_classifier = StackingClassifier(estimators=estimators, final_estimator=final_estimator, cv=5, n_jobs=-1)

# Train the Stacking Classifier
stacking_classifier.fit(X_train, y_train)

# Make predictions using the Stacking Classifier
y_pred_stack = stacking_classifier.predict(X_test)

# Calculate the accuracy of the Stacking Classifier
accuracy_stack = accuracy_score(y_test, y_pred_stack)
print(f"\nStacking Classifier Accuracy: {accuracy_stack:.4f}")

# --- 3. Compare the performance ---
print("\n--- Performance Comparison ---")
print(f"Decision Tree: {accuracy_dt:.4f}")
print(f"SVM: {accuracy_svm:.4f}")
print(f"Logistic Regression: {accuracy_lr:.4f}")
print(f"Stacking Classifier: {accuracy_stack:.4f}")


```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import cross_val_score, KFold, cross_validate
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_wine # Using the Wine dataset
import matplotlib.pyplot as plt

# 1. Load the Dataset (Wine Dataset)
data = load_wine()
X = data.data
y = data.target

print(f"Dataset loaded: Wine Dataset")
print(f"Number of samples: {X.shape[0]}") # Corrected index for number of samples
print(f"Number of features: {X.shape[1]}") # Corrected index for number of features
print(f"Number of classes: {len(np.unique(y))}")

# 2. Initialize Base Estimator
# Decision Trees are a common choice for Bagging due to their tendency to overfit, which Bagging helps to reduce.
base_estimator = DecisionTreeClassifier(random_state=42)

# 3. Initialize Bagging Classifier
# n_estimators: The number of base estimators (decision trees in this case) in the ensemble.
# random_state: Ensures reproducibility of the results.
bagging_classifier = BaggingClassifier(estimator=base_estimator, n_estimators=100, random_state=42)

# 4. Define Cross-Validation Strategy
# KFold: Divides the dataset into K folds, ensuring each fold is used as a test set.
# n_splits=5: Specifies 5-fold cross-validation.
# shuffle=True: Shuffles the data before splitting into folds, helping to prevent bias if the data has an inherent order.
# random_state: Ensures reproducibility of the shuffling.
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# 5. Evaluate the Bagging Classifier using Cross-Validation
# cross_val_score: Returns an array of scores (e.g., accuracy) for each fold of the cross-validation.
# scoring='accuracy': Specifies accuracy as the evaluation metric.
cv_scores = cross_val_score(bagging_classifier, X, y, cv=cv, scoring='accuracy')

print("\nCross-validation scores (accuracy for each fold):", cv_scores)
print(f"Mean CV Accuracy: {np.mean(cv_scores):.4f}")
print(f"Standard Deviation of CV Accuracy: {np.std(cv_scores):.4f}")

# 6. Detailed Cross-Validation with Multiple Metrics (Optional)
# cross_validate: Provides more detailed results, including train scores and fit/score times.
# scoring: A list of metrics to evaluate (accuracy, precision, recall, f1-score).
scoring = ['accuracy', 'precision_macro', 'recall_macro', 'f1_macro']
results = cross_validate(bagging_classifier, X, y, cv=cv, scoring=scoring, return_train_score=True)

print("\n--- Detailed Cross-Validation Results ---")
for metric in scoring:
    print(f"Mean Test {metric}: {np.mean(results[f'test_{metric}']):.4f}")
    print(f"Mean Train {metric}: {np.mean(results[f'train_{metric}']):.4f}")

# 7. Visualize Cross-Validation Scores
plt.figure(figsize=(8, 6))
plt.boxplot(cv_scores)
plt.title('Bagging Classifier Cross-Validation Accuracy Scores (Wine Dataset)')
plt.ylabel('Accuracy Score')
plt.xlabel('Folds')
plt.xticks([]) # Hide x-axis ticks for a single boxplot
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_recall_curve, PrecisionRecallDisplay, average_precision_score
from sklearn.datasets import load_wine # Using the Wine dataset

# 1. Load the Dataset (Wine Dataset, converted to binary)
data = load_wine()
X = data.data
y = data.target

# Create a binary classification problem: classify wine as class 0 or not class 0
# Treat class 0 as the positive class for the PR curve
y_binary = (y == 0).astype(int)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.3, random_state=42)

print(f"Dataset loaded: Wine Dataset (converted to binary)")
print(f"Number of samples: {X.shape}")
print(f"Number of features: {X.shape}")
print(f"Number of binary classes: {len(np.unique(y_binary))}")
print(f"Distribution of classes in binary target (0: Negative, 1: Positive): {np.bincount(y_binary)}")

# 2. Train the Random Forest Classifier
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42) # n_estimators is the number of trees.
rf_classifier.fit(X_train, y_train)

print("\nRandom Forest Classifier trained successfully.")

# 3. Generate Predictions and Plot the Precision-Recall Curve
# Get predicted class probabilities for the positive class (class 1)
# Precision-Recall curves use prediction probabilities rather than hard predictions.
y_pred_prob = rf_classifier.predict_proba(X_test)[:, 1]

# Compute Precision-Recall curve values
# {Link: precision_recall_curve https://scikit-learn.org/stable/modules/generated/sklearn.metrics.precision_recall_curve.html} calculates precision and recall for various thresholds.
precision, recall, thresholds = precision_recall_curve(y_test, y_pred_prob)

# Calculate the Area Under the Precision-Recall Curve (AUPRC) or Average Precision (AP)
# This provides a single score summarizing the curve.
auc_pr = average_precision_score(y_test, y_pred_prob)

print(f"\nArea Under Precision-Recall Curve (AUC-PR): {auc_pr:.4f}")

# Plot the Precision-Recall curve using {Link: PrecisionRecallDisplay https://scikit-learn.org/stable/modules/generated/sklearn.metrics.auc.html}
plt.figure(figsize=(8, 6))
PrecisionRecallDisplay(precision=precision, recall=recall).plot(ax=plt.gca(), name=f'Random Forest (AUC-PR = {auc_pr:.2f})')
plt.title('Precision-Recall Curve for Random Forest Classifier')
plt.xlabel('Recall')
plt.ylabel('Precision')

# Plot the baseline (random classifier)
# The baseline is equal to the prevalence of the positive class.
no_skill = len(y_test[y_test==1]) / len(y_test)
plt.plot([0, 1], [no_skill, no_skill], linestyle='--', label=f'Baseline (Prevalence = {no_skill:.2f})', color='gray')

plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.datasets import load_wine # Using the Wine dataset
import matplotlib.pyplot as plt

# 1. Load the Dataset
data = load_wine()
X = data.data
y = data.target

print(f"Dataset loaded: Wine Dataset")
print(f"Number of samples: {X.shape}") # Corrected index for number of samples
print(f"Number of features: {X.shape}") # Corrected index for number of features
print(f"Number of classes: {len(np.unique(y))}")

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print("\nData split into training and test sets.")

# 2. Bagging with Random Forest Classifier as Base Estimator

# Initialize Random Forest Classifier as the base estimator
# Using a smaller n_estimators for the base RF to keep it somewhat 'weak' as is common for bagging base estimators.
# max_features='sqrt' is a good default for Random Forests.
rf_base_estimator = RandomForestClassifier(n_estimators=10, random_state=42, max_features='sqrt')

# Initialize Bagging Classifier with Random Forest as base
# n_estimators: Number of Random Forest base estimators in the Bagging ensemble.
# bootstrap=True: Samples are drawn with replacement (default for Bagging).
# n_jobs=-1: Use all available CPU cores for parallel processing.
bagging_rf_classifier = BaggingClassifier(
    estimator=rf_base_estimator,
    n_estimators=10,
    random_state=42,
    bootstrap=True,
    n_jobs=-1
)

# Train the model
bagging_rf_classifier.fit(X_train, y_train)

# Make predictions and evaluate accuracy
y_pred_bagging_rf = bagging_rf_classifier.predict(X_test)
accuracy_bagging_rf = accuracy_score(y_test, y_pred_bagging_rf)

print("\n--- Bagging with Random Forest as Base Estimator ---")
print(f"Accuracy: {accuracy_bagging_rf:.4f}")

# 3. Bagging with Logistic Regression Classifier as Base Estimator

# Initialize Logistic Regression Classifier as the base estimator
# Scaling is recommended for Logistic Regression. make_pipeline helps combine scaling and LR.
# solver='liblinear' is a good choice for smaller datasets or for binary classification.
# max_iter: Set explicitly to avoid convergence warnings.
lr_base_estimator = make_pipeline(StandardScaler(), LogisticRegression(random_state=42, solver='liblinear', max_iter=200))

# Initialize Bagging Classifier with Logistic Regression as base
# n_estimators: Number of Logistic Regression base estimators in the Bagging ensemble.
bagging_lr_classifier = BaggingClassifier(
    estimator=lr_base_estimator,
    n_estimators=100,
    random_state=42,
    bootstrap=True,
    n_jobs=-1
)

# Train the model
bagging_lr_classifier.fit(X_train, y_train)

# Make predictions and evaluate accuracy
y_pred_bagging_lr = bagging_lr_classifier.predict(X_test)
accuracy_bagging_lr = accuracy_score(y_test, y_pred_bagging_lr)

print("\n--- Bagging with Logistic Regression as Base Estimator ---")
print(f"Accuracy: {accuracy_bagging_lr:.4f}")

# 4. Comparing Performance
models = ['Bagging (Random Forest)', 'Bagging (Logistic Regression)']
accuracies = [accuracy_bagging_rf, accuracy_bagging_lr]

# Plot the accuracy comparison
plt.figure(figsize=(8, 6))
plt.bar(models, accuracies, color=['skyblue', 'lightcoral'])
plt.ylabel('Accuracy')
plt.title('Comparison of Bagging Classifiers Accuracy (Wine Dataset)')
plt.ylim(0, 1) # Accuracy range from 0 to 1
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print("\n--- Accuracy Comparison Summary ---")
print(f"Bagging with Random Forest Base Estimator Accuracy: {accuracy_bagging_rf:.4f}")
print(f"Bagging with Logistic Regression Base Estimator Accuracy: {accuracy_bagging_lr:.4f}")

if accuracy_bagging_rf > accuracy_bagging_lr:
    print("\nConclusion: Bagging with Random Forest base estimator achieved higher accuracy on the Wine dataset.")
elif accuracy_bagging_lr > accuracy_bagging_rf:
    print("\nConclusion: Bagging with Logistic Regression base estimator achieved higher accuracy on the Wine dataset.")
else:
    print("\nConclusion: Both Bagging classifiers achieved similar accuracy on the Wine dataset.")


```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.datasets import make_regression

# 1. Generate a synthetic regression dataset
X, y = make_regression(n_samples=1000, n_features=10, n_informative=5, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Dataset generated and split.")
print(f"Number of training samples: {X_train.shape}")
print(f"Number of test samples: {X_test.shape}")

# 2. Define different max_samples levels to test
# These represent the proportion of the training data used for each base estimator
max_samples_levels = [0.1, 0.3, 0.5, 0.7, 1.0]

mse_scores = []
r2_scores = []

# Base estimator (Decision Tree Regressor is a common choice for Bagging)
base_regressor = DecisionTreeRegressor(random_state=42)

print("\n--- Training Bagging Regressors with varying max_samples ---")
for max_samples in max_samples_levels:
    print(f"\nTraining with max_samples = {max_samples}")

    # Initialize BaggingRegressor
    # n_estimators: The number of base estimators in the ensemble.
    # max_samples: The number of samples to draw from X to train each base estimator.
    # bootstrap=True: Samples are drawn with replacement (default behavior).
    # random_state: For reproducibility.
    # n_jobs=-1: Use all available CPU cores for parallel processing.
    bagging_regressor = BaggingRegressor(
        estimator=base_regressor,
        n_estimators=100,
        max_samples=max_samples,
        bootstrap=True,
        random_state=42,
        n_jobs=-1
    )

    # Train the model
    bagging_regressor.fit(X_train, y_train)

    # Make predictions
    y_pred = bagging_regressor.predict(X_test)

    # Evaluate performance
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    mse_scores.append(mse)
    r2_scores.append(r2)

    print(f"  Mean Squared Error (MSE): {mse:.4f}")
    print(f"  R-squared (R2): {r2:.4f}")

# 3. Visualizing the performance comparison
plt.figure(figsize=(12, 6))

# Plot MSE
plt.subplot(1, 2, 1)
plt.plot(max_samples_levels, mse_scores, marker='o', linestyle='-', color='skyblue')
plt.title('Bagging Regressor Performance (MSE) vs. max_samples')
plt.xlabel('max_samples (Proportion of Bootstrap Samples)')
plt.ylabel('Mean Squared Error (MSE)')
plt.xticks(max_samples_levels)
plt.grid(True)

# Plot R2
plt.subplot(1, 2, 2)
plt.plot(max_samples_levels, r2_scores, marker='o', linestyle='-', color='lightcoral')
plt.title('Bagging Regressor Performance (R-squared) vs. max_samples')
plt.xlabel('max_samples (Proportion of Bootstrap Samples)')
plt.ylabel('R-squared (R2)')
plt.xticks(max_samples_levels)
plt.grid(True)

plt.tight_layout()
plt.show()

# 4. Summarizing the results
print("\n--- Summary of Performance Comparison ---")
for i, max_samples in enumerate(max_samples_levels):
    print(f"max_samples = {max_samples}: MSE = {mse_scores[i]:.4f}, R2 = {r2_scores[i]:.4f}")

best_mse_index = np.argmin(mse_scores)
print(f"\nBest performance (lowest MSE) achieved with max_samples = {max_samples_levels[best_mse_index]:.1f}")
print(f"Corresponding MSE: {mse_scores[best_mse_index]:.4f}, R2: {r2_scores[best_mse_index]:.4f}")



```
